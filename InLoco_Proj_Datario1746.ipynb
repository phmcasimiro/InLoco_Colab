```python
# Criar atributos/colunas e inserir valores

df_1746.insert(3, 'pais', 'BR', allow_duplicates=False)
df_1746.insert(4, 'uf', 'RJ', allow_duplicates=False)
df_1746.insert(5, 'cd_uf', '33', allow_duplicates=False)
df_1746.insert(6, 'municipio', 'Rio de Janeiro', allow_duplicates=False)
df_1746.insert(7, 'cd_mun6', '330455', allow_duplicates=False)
df_1746.insert(8, 'cd_mun7', '3304557', allow_duplicates=False)
```

```python
# Criar atributos/colunas para quebrar os atributos/colunas DateTime em Ano, Mês e Dia
datetimes_inicio = pd.to_datetime(df_1746['data_inicio'])
# Criar e preencher as novas colunas
df_1746['dia_inicio'] = datetimes_inicio.dt.day
df_1746['mes_inicio'] = datetimes_inicio.dt.month
df_1746['ano_inicio'] = datetimes_inicio.dt.year
```

```python
# Modificar os tipos de dados de alguns atributos
#df_1746['data_inicio'] = df_1746['data_inicio'].astype(str)
#df_1746['data_fim'] = df_1746['data_fim'].astype(str)
#df_1746['data_alvo_finalizacao'] = df_1746['data_alvo_finalizacao'].astype(str)
#df_1746['data_alvo_diagnostico'] = df_1746['data_alvo_diagnostico'].astype(str)
#df_1746['data_real_diagnostico'] = df_1746['data_real_diagnostico'].astype(str)
#df_1746['data_particao'] = df_1746['data_particao'].astype(str)
df_1746['longitude'] = df_1746['longitude'].astype(str)
df_1746['latitude'] = df_1746['latitude'].astype(str)
```

```python
# Transformar um DataFrame em um GeoDataFrame e atribuir um CRS
gdf_1746 = gpd.GeoDataFrame(
    df_1746.loc[:, [c for c in df_1746.columns if c != "geometry"]],
    geometry=gpd.GeoSeries.from_wkt(df_1746["geometry"]),
    crs="epsg:4326",
)
```

```python
# Verificar o CRS
print(gdf_1746.crs)
```

```python
# Verificar o GeoDataFrame
gdf_1746
```

```python
# Filtrar registros com geometrias válidas
gdf_1746_geom = gdf_1746[gdf_1746['geometry'].is_valid]
```

```python
# Filtrar geometrias com valores "nan", ou seja, sem geometria para relizar a geocodificação
gdf_1746_nogeom = gdf_1746[~gdf_1746['geometry'].is_valid]
# este dataframe não será trabalhado neste código
```

```python
# Criar um mapa da folium. 
# Definir localização inicial do mapa (location), nível de zoom (zoom), escala de visibilidade (control_scale)
PCRJ_Map = folium.Map(location=[-22.929010, -43.383911], tiles = 'Stamen Terrain', zoom_start=12, control_scale=True, width='90%', height='90%')

# Adicionar camadas mapa base
folium.TileLayer('openstreetmap').add_to(PCRJ_Map)
folium.TileLayer('stamenterrain', attr="stamenterrain").add_to(PCRJ_Map)
folium.TileLayer('cartodb positron', attr="cartodb positron").add_to(PCRJ_Map)

# Criar um "folium.FeatureGroup" com um laço de repetição do tipo "for" para plotar o GeoDataFrame 'bpm_erj'
Reclamacoes1746_group = folium.FeatureGroup('Reclamações 1746')
for index, reclamacao in gdf_1746_geom.iterrows():

  label_1746_tipo = reclamacao['tipo']
  tooltip_1746 = reclamacao['subtipo']

  folium.Marker(
      [reclamacao.geometry.y, reclamacao.geometry.x],
      icon = folium.Icon(color='black', 
                         icon='phone', 
                         prefix='fa'),
      popup=folium.Popup(html = "Tipo de Reclamação: " + label_1746_tipo + """<img src="https://th.bing.com/th/id/R.4cdf83f09fa4e60003fc40d657faa40b?rik=jsh8xTOzAIF6aA&pid=ImgRaw&r=0" alt="Trulli" style="max-width:100%;max-height:100%">""", max_width=450),
      tooltip = tooltip_1746
      ).add_to(PCRJ_Map)
Reclamacoes1746_group.add_to(PCRJ_Map)

# Adicionar controle de camadas no mapa
folium.LayerControl().add_to(PCRJ_Map)
```
